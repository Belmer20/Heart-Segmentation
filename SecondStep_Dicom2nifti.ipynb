```python
import os
from glob import glob
import SimpleITK as sitk

# Output paths
out_path_images = "C:/Users/TRETEC/Desktop/Segmenetation/Task02_Heart/nifti_files/images"
out_path_labels = "C:/Users/TRETEC/Desktop/Segmenetation/Task02_Heart/nifti_files/labels"

# Input paths
in_path_images = "C:/Users/TRETEC/Desktop/Segmenetation/Task02_Heart/dicom_groups/images/*"
in_path_labels = "C:/Users/TRETEC/Desktop/Segmenetation/Task02_Heart/dicom_groups/labels/*"

# Create the output directories if they do not exist
os.makedirs(out_path_images, exist_ok=True)
os.makedirs(out_path_labels, exist_ok=True)

# Get the list of paths for the DICOM series
list_images = glob(in_path_images)
list_labels = glob(in_path_labels)

# Convert DICOM to NIfTI for images
for patient in list_images:
    patient_name = os.path.basename(os.path.normpath(patient))
    output_file = os.path.join(out_path_images, patient_name + '.nii.gz')

    # Read the DICOM series
    reader = sitk.ImageSeriesReader()
    dicom_series = reader.GetGDCMSeriesFileNames(patient)
    reader.SetFileNames(dicom_series)

    # Convert the series to an image
    image = reader.Execute()

    # Save the image as a NIfTI file
    sitk.WriteImage(image, output_file)
    print(f'Converted {patient} to {output_file}')

# Convert DICOM to NIfTI for labels
for patient in list_labels:
    patient_name = os.path.basename(os.path.normpath(patient))
    output_file = os.path.join(out_path_labels, patient_name + '.nii.gz')

    # Read the DICOM series
    reader = sitk.ImageSeriesReader()
    dicom_series = reader.GetGDCMSeriesFileNames(patient)
    reader.SetFileNames(dicom_series)

    # Convert the series to an image
    image = reader.Execute()

    # Save the image as a NIfTI file
    sitk.WriteImage(image, output_file)
    print(f'Converted {patient} to {output_file}')
```
